```python
def floor(num):
    return round(num-0.5)

def getBin(num, size):
    b = bin(num)[2:]
    return (size-len(b))*'0' + b

class binaryTree:
    def __init__(self, data, heap = False): # Initializes the binary tree from a python list
        self.heap = heap
        self.data = data
        if type(self.data) == int or type(self.data) == float:
            self.head = binaryTreeNode(self.data)
        elif type(self.data) == list:
            if self.heap == True:
                self.data.sort()
            self.head = binaryTreeNode(data[0])
            for i in range(1,len(data)):
                self.insert(data[i])
        else:
            raise Exception("Please enter either a number or list of numbers")
        
    def lookup(self, value, showPath = True): # Return a node in the tree and the path to get there
        current = self.head
        path = ''
        point = ''
        while self.heap = False:
            try:
                if value == current.value:
                    path += point + str(current.value)
                    break
                elif value < current.value:
                    path += point + str(current.value)
                    current = current.lChild
                    point = ' ↘ '
                else:
                    path += point + str(current.value)
                    current = current.rChild
                    point = ' ↗ '
            except AttributeError:
                path += point + 'X'
                current = -1
                break
        
        while self.heap == True:
            
        
        if showPath == True:
            print(path)
        return current
    
    def insert(self, value):
        current = self.head
        while True:
            if value < current.value:
                if current.lChild == None:
                    current.lChild = binaryTreeNode(value)
                    break
                else:
                    current = current.lChild
            else:
                if current.rChild == None:
                    current.rChild = binaryTreeNode(value)
                    break
                else:
                    current = current.rChild
    
    def display(self, start = None):
        if start == None:
            start = self.head
        lines, *_ = start._display_aux()
        for line in lines:
            print(line)
            
    def getLayer(self, layer, keepNull = False):
        c = 0
        size = 2**layer-1
        layerList = []
        while c <= size:
            current = self.head
            num = getBin(c,layer)
            for bit in num:
                try:
                    if bit == '0':
                        current = current.lChild
                    else:
                        current = current.rChild
                except AttributeError:
                    continue
            if ~(keepNull == False and current == None):
                layerList.append(current)
            c += 1
        return layerList
                    
class binaryTreeNode: # Class for each node in a binary tree
    def __init__(self, value, leftChild = None, rightChild = None):
        self.value = value
        self.lChild = leftChild
        self.rChild = rightChild
     
    # The following function was repurposed from user J. V. at https://stackoverflow.com/questions/34012886/print-binary-tree-level-by-level-in-python
    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.rChild is None and self.lChild is None:
            line = '%s' % self.value
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.rChild is None:
            lines, n, p, x = self.lChild._display_aux()
            s = '%s' % self.value
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.lChild is None:
            lines, n, p, x = self.rChild._display_aux()
            s = '%s' % self.value
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.lChild._display_aux()
        right, m, q, y = self.rChild._display_aux()
        s = '%s' % self.value
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2
```

```python
import random
randTree = binaryTree(50)
c = 0
while c < 75: # While loop that inserts 75 random unique numbers into the tree
    i = random.randint(0,100)
    if randTree.lookup(i, False) == -1: # Only insert a random number if it doesn't already exist in the tree
        randTree.insert(i)
        c += 1
randTree.display()
```
